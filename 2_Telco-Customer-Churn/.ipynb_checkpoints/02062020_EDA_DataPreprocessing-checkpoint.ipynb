# 1) Data Preprocessing

```python
df['SeniorCitizen'] = df['SeniorCitizen'].replace({0: 'No', 1: 'Yes'})

df['MultipleLines'] = df['MultipleLines'].replace({'No phone service': 'No'})

replace_cols = ['OnlineSecurity', 
                'OnlineBackup', 
                'DeviceProtection', 
                'TechSupport', 
                'StreamingTV', 
                'StreamingMovies']
for col in replace_cols:
    df[col] = df[col].replace({'No internet service': 'No'})

df['MonthlyCharges'] = df['MonthlyCharges'].astype(float)
df['TotalCharges'] = df['TotalCharges'].replace(" ", 0).astype(float)
```

```python
display(df.head())
display(df.info())
print ("Unique values:\n", df.nunique())
```

### Numerical, categorical, target column

```python
# Numerical, categorical, target column

target_col = ['Churn']

id_col = ['customerID']

num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']   
cat_cols = [x for x in df.columns if x not in num_cols + target_col + id_col]
# cat_cols = df.nunique()[df.nunique() < 5].keys().tolist()

print('cat_cols:', cat_cols)
```

# 3) Data Preprocessing
* num_col: Standardscaler
* cat_col: OneHotEncoder

```python
from sklearn.preprocessing import StandardScaler

df_scaled = df.copy().drop(id_col, axis=1)

scaler = StandardScaler()
df_scaled[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
# onehot encoded & combined with scaled df 

df_combined = pd.concat([df_scaled,
                         pd.get_dummies(df[cat_cols], drop_first=True)], axis=1).drop(cat_cols, axis=1)

display(df_combined.head())
print(df_combined.shape)
```

# Finally, save preprocessed csv

```python
# Split features & target value from training set

df_combined.to_csv('./input/df_combined.csv', index=False)
```
